```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg.norm as norm
```

```python
v_1 = np.array([2,4,6])
v_2 = np.array([-8,7,0])
v_3 = np.array([1,0,0,1])
```

```python
ps = np.exp(np.arange(-10,3,0.1))

def pnorm(v,p):
    terms = np.exp(p * np.log(np.abs(v)))
    sum = np.sum(terms)
    return np.exp(np.log(sum)/p)

for p in ps:
    print(p, pnorm(v_3,p))
```
